```python
import sys
import cv2
import imp
from PyQt5 import QtCore, QtGui, QtWidgets
from PyQt5.QtCore import *
from PyQt5.QtWidgets import *
from PyQt5.QtGui import *
import os
from MainWindow import Ui_MainWindow
import numpy as np
import math
import pdb

from Snake_GVF import Snake_GVF
from Snake_classical import Snake_classical

envpath = '/home/x/.conda/envs/ML/lib/python3.9/site-packages/cv2/qt/plugins/platforms'
os.environ['QT_QPA_PLATFORM_PLUGIN_PATH'] = envpath


def mkdir(dir):
    if os.path.exists(dir):
        return
    else:
        os.makedirs(dir)

class ClickHandler:
    image = None
    POINTS_SIZE = 0

    def __init__(self, image):
        self.image = image
        cv2.putText(self.image, 'click to draw the initial contour', (50, 50), cv2.FONT_HERSHEY_SIMPLEX, .5, (0, 0, 0), 1)
        cv2.imshow('Snakes', image)

        h, w, _ = self.image.shape
        self.counter = 0
        self.points = []

    def get_points(self):
        return self.points

    def click_event(self, event, x, y, flags, params):
        if event == cv2.EVENT_LBUTTONDOWN:
            point = np.array([x, y])
            cv2.circle(self.image, (x, y), 1, (255, 140, 0), 10)
            cv2.imshow('Snakes', self.image)
            self.points.append(point)


class SnakeWindow(QtWidgets.QMainWindow, Ui_MainWindow):
    def __init__(self):
        super(SnakeWindow, self).__init__()
        self.setupUi(self)
        self.image = None
        self.saveResult = self.SaveResultBox.isChecked()
        self.SelectImageButton.clicked.connect(self.on_click_select_image)
        self.ClassicalSnakeButton.clicked.connect(self.on_click_classical_snake_button)
        self.GVFSnakeButton.clicked.connect(self.on_click_GVF_snake_button)

    def on_click_select_image(self):
        image_path = QtWidgets.QFileDialog.getOpenFileName(self, '选取文件', './', "图像文件 (*.jpg *.jpeg *.png *.bmp *.PNG)")
        
        self.image = cv2.imread(image_path[0], cv2.IMREAD_COLOR)
        if self.image is not None:
            h, w = self.image.shape[0], self.image.shape[1]
            if h > 1000 or w > 1000:
                self.image = cv2.pyrDown(self.image)
            self.copy = self.image.copy()
            self.mouse_handler = ClickHandler(self.image)
            cv2.setMouseCallback('Snakes', self.mouse_handler.click_event)
            cv2.waitKey(0)

    def run_snake(self, snake, snake_type):
        # cv2.destroyAllWindows()
        self.saveResult = self.SaveResultBox.isChecked()
        if self.saveResult:
            mkdir(os.path.join('results', snake_type))
            cv2.imwrite(os.path.join('results', snake_type, 'first_frame.png'), snake.update_frame())

        frame_array = []
        step = 0
        maxIter = self.MaxIterBox.value()
        while step < maxIter:
            new_frame = snake.update_frame()
            if self.saveResult:
                frame_array.append(new_frame)
            cv2.imshow('Snake is running ...', new_frame)
            h, w, _ = new_frame.shape
            snake.step()
            key = cv2.waitKey(100)
            # if key = ESC break loop
            if key == 27:
                break
            step += 1
        
        if self.saveResult:
            cv2.imwrite(os.path.join('results', snake_type, 'result.png'), snake.update_frame())
        
            # instantiate video writer
            fourcc = cv2.VideoWriter_fourcc(*'mp4v')
            out = cv2.VideoWriter(os.path.join('results', snake_type, 'contour.mp4'), fourcc, 15, (w, h))
            for i in range(len(frame_array)):
                out.write(frame_array[i])
            out.release()
        cv2.waitKey(0)

    def on_click_classical_snake_button(self):
        if self.image is None:
            return
        if self.CSDefaultBox.isChecked():
            CSnake = Snake_classical(self.copy, self.mouse_handler.get_points())
            self.run_snake(CSnake, snake_type='Classical Snake')
        else:
            alpha = self.CSAlphaSB.value()
            beta = self.CSBetaSB.value()
            gamma = self.CSGammaSB.value()
            wl = self.CSwlSB.value()
            we = self.CSweSB.value()
            wt = self.CSwtSB.value()
            CSnake = Snake_classical(self.copy, self.mouse_handler.get_points(), alpha, beta, gamma, wl, we, wt)
            self.run_snake(CSnake, snake_type='Classical Snake')

    def on_click_GVF_snake_button(self):
        if self.image is None:
            return
        if self.GVFDefaultBox.isChecked():
            GVFSnake = Snake_GVF(self.copy, self.mouse_handler.get_points())
            self.run_snake(GVFSnake, snake_type='GVF Snake')
        else:
            alpha = self.GVFAlphaSB.value()
            beta = self.GVFBetaSB.value()
            mu = self.GVFMuSB.value()
            GVFSnake = Snake_GVF(self.copy, self.mouse_handler.get_points(), alpha, beta, mu)
            self.run_snake(GVFSnake, snake_type='GVF Snake')


if __name__ == '__main__':
    app = QtCore.QCoreApplication.instance()
    if app is None:
        app = QtWidgets.QApplication(sys.argv)
    window = SnakeWindow()
    window.show()
    sys.exit(app.exec_())
```
